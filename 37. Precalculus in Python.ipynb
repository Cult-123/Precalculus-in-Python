# Plotting 2D function or equation Graph (All Functions are Equations but All Equations are Not Functions)
* By providing 0 as Input value of a function f(X), whatever the Output we will get from the function f(X) will be the Intercept of Y-axis.
* On Y-axis, only One Intercept is Possible

## Identity Function. The equation is $y = f(x) = x$
* Output is exactly same as Input
* Produces straight line diagonal to the 1st & 3rd quadrant passing through the central coordinate (0,0) of X & Y axis.

<img src="DataFile/For any 2D Function.jpg" style="height:250px; width:700px"/>

```python
# plotting a 2D plot, Identity function
# The equation is y = f(x) = x
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10,10,100)
y = x

# Plot the Data
#plt.plot(x, y, label='Identity Function y = x', color='blue', linewidth=2)  # label is interconnected with plt.legend()
#plt.figure(figsize=(8, 6))
plt.plot(x, y, label="Identity Function $y = x$")    # $ $ dollar makes the formula more presentable

# Highlight X and Y axes with thick black lines
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.axvline(0, color='black', linewidth=1, linestyle='--')

# Add titles, labels, grid, legend
plt.title("Plot of $y = f(x) = x$")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True) # Optional: Add a grid for better readability
plt.legend()    # interconnected with plt.plot() label

# Optional: add minor ticks
plt.minorticks_on()

# -------------------------------------------------
# Show the plot
# -------------------------------------------------
plt.tight_layout()   # prevents label cutoff & Overlaps automatically if exist
plt.show()           # It removes hexadecimal number of the memory address
```

## Rational Function. The equation is $y = f(x) = 1 / (X - 1)$

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Define the rational function
def rational_function(x):
    return 1 / (x - 1)

# 2. Split the domain to avoid the vertical asymptote at x = 1
# Generate points to the left of the asymptote
x_left = np.linspace(-4, 0.99, 250)
# Generate points to the right of the asymptote
x_right = np.linspace(1.01, 4, 250)

# Calculate the corresponding y-values for each side
y_left = rational_function(x_left)
y_right = rational_function(x_right)

# 3. Plot the function and asymptotes
plt.figure(figsize=(8, 6))

# Plot the two parts of the curve separately
plt.plot(x_left, y_left, color='blue', linewidth=2, label=r"$f(x) = \frac{1}{x - 1}$")
plt.plot(x_right, y_right, color='blue', linewidth=2)

# Mark the vertical asymptote
plt.axvline(1, color='red', linestyle='--', linewidth=1, label="Vertical Asymptote (x=1)")

# Mark the horizontal asymptote (y=0 in this case)
plt.axhline(0, color='green', linestyle='--', linewidth=1, label="Horizontal Asymptote (y=0)")

# 4. Add labels, a legend, and display the plot
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot of a Rational Function $y = f(x) = 1 / (x - 1)$ with Asymptotes")
plt.xlim([-4, 4])
plt.ylim([-5, 5])
plt.legend()
# Optional: add minor ticks
plt.minorticks_on()
plt.grid(True)
plt.tight_layout()   # prevents label cutoff & Overlaps automatically if exist
plt.show()
```

## Plotting a 2D plot, Parabola function. The equation is $y = f(x) = x^2$
* It's secant Line Equation Between two Points
* **A secant line cuts through a curve, intersecting it at two or more points, representing the average rate of change.**
* **A tangent line touches a curve at exactly one point (locally) and shows the instantaneous rate of change (derivative) at that point, essentially being the limit of secant lines as the two points get closer.**
* **Think of a secant as an extended chord, while a tangent "kisses" the curve at one spot.**

<img src="DataFile/Secant Line Function Equation.jpg"/>

```python
# plotting a 2D plot, Parabola function
# The equation is y = f(x) = x^2
x = np.linspace(-10,10,100)
y = x**2

# 1. Define the parabola function
def parabola(x):
    return x**2

# 2. Define the two points for the secant line
x1, x2 = -2.0, 7.5
y1, y2 = parabola(x1), parabola(x2)

# 3. Calculate the equation of the secant line (y = mx + c)
# Slope (m)
m = (y2 - y1) / (x2 - x1)
# Y-intercept (c)
c = y1 - m * x1

def secant_line(x):
    return m * x + c

# Secant line points, only for the interval between x1 and x2
x_secant = np.linspace(min(x1, x2), max(x1, x2), 2)
y_secant = secant_line(x_secant)

# Plot the Data
plt.plot(x, y, label="$f(x) = x^2$")
plt.plot(x_secant, y_secant, 
         label=f'Secant Line S(x) = {m:.2f}x + {c:.2f} (m is slope = {m:.2f}, c is y-intercept = {c:.2f})', 
         color='red', linestyle='--')
plt.scatter([x1, x2], [y1, y2], color='red', zorder=5, label=f"Points: ({x1}, {y1:.2f}), ({x2}, {y2:.2f})")

# Add titles, labels, grid, legend
plt.title("Plot of Parabola Function $y = f(x) = x^2$")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.minorticks_on()

plt.tight_layout()
plt.show()

print(f"Points on Parabola: ({x1}, {y1}), ({x2}, {y2})")
print(f"Value on the Secant line equation at x=2: {secant_line(2)}")
```

# Solving Quadratic (Polynomial) Equation Functions (Finding X-axis & Y-axis Intercept Points)
* By providing 0 as Input value of a function f(X), whatever the Output we will get from the function f(X) will be the Intercept of Y-axis.
* On Y-axis, only One Intercept is Possible
* On X-axis, Multiple Intercept points are Possible
* For finding X-axis Intercept, f(X) = 0, Make the equation = 0, then solve
* For finding Y-axis Intercept, f(0), Plugin 0 into the equation and then check the Output

## Methods for Solving Quadratic (Polynomial) Equation Functions
* Taking the Square Root Method
* Factoring Method
* By Completing the Square
* By using the Quadratic Formula

```python
from sympy import symbols, solve

# Define the variable
x = symbols('x')

# Define your function
fx = x**2 - 4

print("=" * 50)
print(f"Function: f(x) = {fx}")
print("=" * 50)

# 1. Find X-AXIS INTERCEPTS (where y = 0, i.e., f(x) = 0)
x_intercepts = solve(fx, x)
print(f"\nX-axis intercepts (roots): {x_intercepts}")
print(f"X-axis intercept point Coordinates are: ", end="")
for x_val in x_intercepts:
    print(f"({x_val}, 0)", end=" ")
print()

# 2. Find Y-AXIS INTERCEPT (where x = 0, i.e., f(0))
y_intercept = fx.subs(x, 0)
print(f"\nY-axis intercept: {y_intercept}")
print(f"Y-axis intercept point Coordinate is: (0, {y_intercept})")
print("=" * 50)
```

## More Examples with Different Functions

```python
from sympy import symbols, solve, sqrt

x = symbols('x')

def find_intercepts(fx, function_name=""):
    """Function to find x and y intercepts"""
    print("\n" + "=" * 60)
    print(f"{function_name}")
    print(f"Function: f(x) = {fx}")
    print("=" * 60)
    
    # X-intercepts
    try:
        x_intercepts = solve(fx, x)
        print(f"X-axis intercepts: {x_intercepts}")
        if x_intercepts:
            print(f"Coordinates: ", end="")
            for x_val in x_intercepts:
                print(f"({x_val}, 0)", end=" ")
            print()
        else:
            print("No x-intercepts")
    except Exception as e:
        print(f"Could not solve for x-intercepts: {e}")
    
    # Y-intercept
    try:
        y_intercept = fx.subs(x, 0)
        print(f"\nY-axis intercept: {y_intercept}")
        print(f"Coordinate: (0, {y_intercept})")
    except Exception as e:
        print(f"Could not find y-intercept: {e}")
    
    print("=" * 60)

# Example 1: Quadratic (Parabola)
fx1 = x**2 - 4
find_intercepts(fx1, "Example 1: Quadratic Function")

# Example 2: Linear Function
fx2 = 2*x + 6
find_intercepts(fx2, "Example 2: Linear Function")

# Example 3: Cubic Function
fx3 = x**3 - 6*x**2 + 11*x - 6
find_intercepts(fx3, "Example 3: Cubic Function")

# Example 4: Function with no real x-intercepts
fx4 = x**2 + 4
find_intercepts(fx4, "Example 4: Parabola (no real x-intercepts)")

# Example 5: Rational Function
fx5 = (x**2 - 9) / (x - 2)
find_intercepts(fx5, "Example 5: Rational Function")
```

# Finding Intersections of 2 or more Functions (Finding Common Intersection Point Coordinates)

```python
from sympy import symbols, solve

# Define the variable
x = symbols('x')

# Define two functions
fx = x**2 + 4*x - 3
gx = 2*x + 1

print("=" * 60)
print(f"Function f(x) = {fx}")
print(f"Function g(x) = {gx}")
print("=" * 60)

# Find intersection points by solving f(x) = g(x)
# This is equivalent to solving f(x) - g(x) = 0
intersection_x = solve(fx - gx, x)

print(f"\nIntersection x-coordinates: {intersection_x}")
print("\nIntersection point(s):")

for x_val in intersection_x:
    y_val = fx.subs(x, x_val)  # or gx.subs(x, x_val), should be the same
    print(f"  ({x_val}, {y_val})")
    print(f"  Decimal: ({float(x_val):.4f}, {float(y_val):.4f})")

# Verify both functions give same y-value
print("\nVerification:")
for x_val in intersection_x:
    y_from_f = fx.subs(x, x_val)
    y_from_g = gx.subs(x, x_val)
    print(f"  At x = {x_val}: f(x) = {y_from_f}, g(x) = {y_from_g}")

print("=" * 60)
```

## Multiple Function Pairs of Intersection

```python
from sympy import symbols, solve, simplify

def find_intersection(f1, f2, func1_name="f(x)", func2_name="g(x)"):
    """Find intersection points between two functions"""
    x = symbols('x')
    
    print("\n" + "=" * 60)
    print(f"{func1_name} = {f1}")
    print(f"{func2_name} = {f2}")
    print("-" * 60)
    
    # Solve f1(x) = f2(x)
    intersection_x = solve(f1 - f2, x)
    
    if not intersection_x:
        print("No intersection points found!")
        return []
    
    print(f"Intersection x-values: {intersection_x}")
    print("\nIntersection Points:")
    
    points = []
    for x_val in intersection_x:
        y_val = f1.subs(x, x_val)
        y_val = simplify(y_val)
        points.append((x_val, y_val))
        
        print(f"  Exact: ({x_val}, {y_val})")
        try:
            print(f"  Decimal: ({float(x_val):.4f}, {float(y_val):.4f})")
        except:
            print(f"  (Complex or non-numeric solution)")
    
    print("=" * 60)
    return points

# Define the variable
x = symbols('x')

# Example 1: Parabola and Line
print("\n*** EXAMPLE 1: Parabola and Line ***")
fx = x**2 + 4*x - 3
gx = 2*x + 1
find_intersection(fx, gx, "f(x)", "g(x)")

# Example 2: Two Parabolas
print("\n*** EXAMPLE 2: Two Parabolas ***")
fx = x**2 - 4
gx = -x**2 + 2*x + 4
find_intersection(fx, gx, "f(x)", "g(x)")

# Example 3: Parabola and Cubic
print("\n*** EXAMPLE 3: Parabola and Cubic ***")
fx = x**2
gx = x**3 - 2*x
find_intersection(fx, gx, "f(x)", "g(x)")

# Example 4: Two Lines
print("\n*** EXAMPLE 4: Two Lines ***")
fx = 3*x + 5
gx = -2*x + 10
find_intersection(fx, gx, "f(x)", "g(x)")
```

## Finding ALL Pairwise Intersections (3+ Functions)

```python
from sympy import symbols, solve, simplify
from itertools import combinations

x = symbols('x')

# Define multiple functions
functions = {
    'f(x)': x**2 - 4,
    'g(x)': 2*x,
    'h(x)': -x + 3
}

print("=" * 70)
print("FINDING ALL PAIRWISE INTERSECTIONS")
print("=" * 70)

for name in functions:
    print(f"{name} = {functions[name]}")

print("=" * 70)

# Find all pairwise intersections
for (name1, func1), (name2, func2) in combinations(functions.items(), 2):
    print(f"\n*** Intersection of {name1} and {name2} ***")
    
    intersection_x = solve(func1 - func2, x)
    
    if not intersection_x:
        print("  No intersection points")
        continue
    
    for x_val in intersection_x:
        y_val = func1.subs(x, x_val)
        y_val = simplify(y_val)
        print(f"  Point: ({x_val}, {y_val})", end="")
        try:
            print(f"  ({float(x_val):.4f}, {float(y_val):.4f})")
        except:
            print()

print("\n" + "=" * 70)
```

## Complete Inequality Solver with Interpretation

```python
from sympy import symbols, solve, solve_univariate_inequality, Interval

def solve_quadratic_inequality(fx, inequality_type=">"):
    """
    Solve quadratic inequality
    inequality_type: '>', '>=', '<', '<='
    """
    x = symbols('x', real=True)
    
    print("\n" + "=" * 70)
    print(f"Function: f(x) = {fx}")
    print(f"Inequality condition: f(x) {inequality_type} 0")
    print("-" * 70)
    
    # Create the inequality
    if inequality_type == '>':
        inequality = fx > 0
    elif inequality_type == '>=':
        inequality = fx >= 0
    elif inequality_type == '<':
        inequality = fx < 0
    elif inequality_type == '<=':
        inequality = fx <= 0
    
    # Solve the inequality
    solution = solve_univariate_inequality(inequality, x, relational=False)
    
    print(f"Condition Solution: {solution}")
    
    # Find the roots (where f(x) = 0)
    roots = solve(fx, x)
    print(f"Roots (f(x) = 0): {roots}")
    
    # Interpret the solution
    if hasattr(solution, 'as_relational'):
        print(f"Interval notation: {solution}")
    
    return solution

# Define variable
x = symbols('x', real=True)

# Example 1: x^2 - 5x + 6 (opens upward, two real roots)
print("\n*** EXAMPLE 1: x - 5x + 6 ***")
fx1 = x**2 - 5*x + 6
solve_quadratic_inequality(fx1, '>')
solve_quadratic_inequality(fx1, '<')

# Example 2: -x^2 + 4x - 3 (opens downward)
print("\n*** EXAMPLE 2: -x + 4x - 3 ***")
fx2 = -x**2 + 4*x - 3
solve_quadratic_inequality(fx2, '>')
solve_quadratic_inequality(fx2, '<')

# Example 3: x^2 + 2x + 5 (no real roots)
print("\n*** EXAMPLE 3: x + 2x + 5 ***")
fx3 = x**2 + 2*x + 5
solve_quadratic_inequality(fx3, '>')
solve_quadratic_inequality(fx3, '<')

# Example 4: x^2 - 4x + 4 (one repeated root)
print("\n*** EXAMPLE 4: x - 4x + 4 ***")
fx4 = x**2 - 4*x + 4
solve_quadratic_inequality(fx4, '>=')
solve_quadratic_inequality(fx4, '<')
```

## Comprehensive Inequality Analyzer

```python
from sympy import symbols, solve, factor, expand

def analyze_quadratic_inequality(fx):
    """Complete analysis of quadratic inequality"""
    x = symbols('x', real=True)
    
    print("\n" + "=" * 70)
    print(f"QUADRATIC INEQUALITY ANALYSIS")
    print("=" * 70)
    print(f"Function: f(x) = {fx}")
    print(f"Factored form: f(x) = {factor(fx)}")
    print("-" * 70)
    
    # Find coefficients
    expanded = expand(fx)
    coeffs = [expanded.coeff(x, 2), expanded.coeff(x, 1), expanded.as_coeff_add(x)[0]]
    a, b, c = coeffs
    
    print(f"Standard form: f(x) = {a}x + {b}x + {c}")
    print(f"Coefficient a = {a}  Parabola opens {'upward' if a > 0 else 'downward'}")
    
    # Find discriminant
    discriminant = b**2 - 4*a*c
    print(f"Discriminant () = b - 4ac = {discriminant}")
    
    # Find roots
    roots = solve(fx, x)
    print(f"Roots: {roots}")
    
    if len(roots) == 0:
        print("   No real roots (parabola doesn't cross x-axis)")
    elif len(roots) == 1:
        print(f"   One repeated root at x = {roots[0]}")
    else:
        print(f"   Two distinct roots at x = {roots[0]} and x = {roots[1]}")
    
    print("\n" + "-" * 70)
    print("INEQUALITY SOLUTIONS:")
    print("-" * 70)
    
    # Solve all four types of inequalities
    for ineq_type, ineq_symbol in [('>', '>'), ('>=', ''), ('<', '<'), ('<=', '')]:
        if ineq_type == '>':
            solution = solve(fx > 0, x)
        elif ineq_type == '>=':
            solution = solve(fx >= 0, x)
        elif ineq_type == '<':
            solution = solve(fx < 0, x)
        elif ineq_type == '<=':
            solution = solve(fx <= 0, x)
        
        print(f"f(x) {ineq_symbol} 0: {solution}")
    
    print("=" * 70)

# Define variable
x = symbols('x', real=True)

# Test different cases
print("\n*** CASE 1: Two distinct real roots, opens upward ***")
analyze_quadratic_inequality(x**2 - 5*x + 6)

print("\n*** CASE 2: Two distinct real roots, opens downward ***")
analyze_quadratic_inequality(-x**2 + 4*x - 3)

print("\n*** CASE 3: No real roots, opens upward ***")
analyze_quadratic_inequality(x**2 + 2*x + 5)

print("\n*** CASE 4: One repeated root ***")
analyze_quadratic_inequality(x**2 - 4*x + 4)
```

## Plotting Inequality Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
from sympy import symbols, solve, lambdify

def plot_quadratic_inequality(fx, inequality_type='>', y_limit=10):
    """Visualize quadratic inequality"""
    x = symbols('x')
    
    # Find roots
    roots = solve(fx, x)
    roots_float = [float(r) for r in roots if r.is_real]
    
    # Solve inequality
    if inequality_type == '>':
        solution = solve(fx > 0, x)
        title_ineq = 'f(x) > 0'
    elif inequality_type == '>=':
        solution = solve(fx >= 0, x)
        title_ineq = 'f(x)  0'
    elif inequality_type == '<':
        solution = solve(fx < 0, x)
        title_ineq = 'f(x) < 0'
    elif inequality_type == '<=':
        solution = solve(fx <= 0, x)
        title_ineq = 'f(x)  0'
    
    # Convert to numpy function
    f_np = lambdify(x, fx, 'numpy')
    
    # Create x values
    if roots_float:
        margin = 2
        x_min = min(roots_float) - margin
        x_max = max(roots_float) + margin
    else:
        x_min, x_max = -5, 5
    
    x_vals = np.linspace(x_min, x_max, 1000)
    y_vals = f_np(x_vals)
    
    # Create figure
    fig, ax = plt.subplots(figsize=(12, 7))
    
    # Plot the function
    ax.plot(x_vals, y_vals, 'b-', linewidth=2.5, label=f'$f(x) = {smp.latex(fx)}$')
    
    # Shade the solution region
    if inequality_type in ['>', '>=']:
        ax.fill_between(x_vals, y_vals, y_limit, where=(y_vals >= 0), 
                        alpha=0.3, color='green', label='Solution region')
    else:
        ax.fill_between(x_vals, y_vals, -y_limit, where=(y_vals <= 0), 
                        alpha=0.3, color='red', label='Solution region')
    
    # Plot roots
    if roots_float:
        ax.scatter(roots_float, [0]*len(roots_float), color='red', 
                  s=150, zorder=5, marker='o', edgecolors='black', linewidths=2)
        for r in roots_float:
            ax.annotate(f'x = {r:.2f}', xy=(r, 0), xytext=(10, 20),
                       textcoords='offset points', fontsize=11,
                       bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.8),
                       arrowprops=dict(arrowstyle='->', color='black'))
    
    # Axes and grid
    ax.axhline(y=0, color='black', linewidth=1.5)
    ax.axvline(x=0, color='black', linewidth=1.5)
    ax.grid(True, alpha=0.3)
    
    # Labels
    ax.set_xlabel('x', fontsize=13)
    ax.set_ylabel('f(x)', fontsize=13)
    ax.set_title(f'Quadratic Inequality: {title_ineq}\nSolution: {solution}', 
                fontsize=14, fontweight='bold')
    ax.legend(fontsize=11, loc='best')
    ax.set_ylim(-y_limit, y_limit)
    
    plt.tight_layout()
    plt.show()
    
    # Print solution
    print("=" * 70)
    print(f"Function: f(x) = {fx}")
    print(f"Inequality: {title_ineq}")
    print(f"Roots: {roots}")
    print(f"Solution: {solution}")
    print("=" * 70)

# Examples
x = symbols('x')

# Example 1: x - 5x + 6 > 0
plot_quadratic_inequality(x**2 - 5*x + 6, '>')

# Example 2: x - 5x + 6 < 0
plot_quadratic_inequality(x**2 - 5*x + 6, '<')

# Example 3: -x + 4x - 3 > 0
plot_quadratic_inequality(-x**2 + 4*x - 3, '>')

# Example 4: x + 2x + 5 > 0 (no real roots)
plot_quadratic_inequality(x**2 + 2*x + 5, '>')
```

```python
from sympy import symbols, solve
import sympy as smp

# Define the variable
x = symbols('x')

# Per unit Selling price Function Equation. y intercept(b) = 100, Slope m = -(1/3)
# This is Linear Equation
# X Input will be total number of Unit
# Y Output will be price of Each Unit
Px = -(1/3)*x + 100


# Revenue Function Equation. a = -(1/3), b = 100
# This is Downward Opening Parabola
# X Input will be total number of Unit
# Y Output will be Revenue
Rx = -(1/3)*x**2 + 100*x   # Rx = x * Px


# Tangent Slope Function (Derivative) of the Rx Function
dRx = smp.diff(Rx, x)



# Printing the Functions
print(f"Price per unit function: P(x) = {Px}")
print(f"Total revenue function: R(x) = {Rx}")
print(f"Tangent Slope Function (Derivative) of R(x): dR(x) = {dRx}")
print()
print("Where x = number of units as Input for both P(x) & R(x)")
print("Where x = Specific point x-coordinate, number of units as Input for Tangent Slope Function dR(x)")


print("=" * 70)
print("SOLVING QUADRATIC INEQUALITIES")
print("=" * 70)


# 2. Solve R(x) >= 0
print("\n1. Inequalities Condition R(x) >= 0:")
solution1 = solve(Rx >= 0, x)
print(f"   Condition Solution: {solution1}")

print("\n" + "="*70)
print("STEP 2: FINDING VERTEX WHERE TANGENT LINE SLOPE = 0")
print("="*70)
print(f"Tangent Slope Function (Derivative) of R(x): dR(x) = {dRx}")


# 1. Find X-AXIS INTERCEPTS (where y = 0, slope m =0, i.e., dR(x) = 0)
x_intercepts = solve(dRx, x)
print(f"\nX-axis intercepts (roots): {x_intercepts}")
print(f"X-axis intercept point Coordinates are: ", end="")
for x_val in x_intercepts:
    print(f"({x_val}, 0)", end=" ")
print()


# Defining the functions seperately
def Px(x):
    return -(1/3)*x + 100

def Rx(x):
    return -(1/3)*x**2 + 100*x

def dRx(x):
    return 100 - 0.666666666666667*x

# ============================================================================
# FINAL ANSWER
# ============================================================================
print("\n" + "="*70)
print(" FINAL ANSWER: MAXIMUM REVENUE CONFIGURATION")
print("="*70)
print(f"""
To MAXIMIZE the total revenue:

    Total Number of Units to Produce/Sell: {x_intercepts[0]:.2f} units
   
    Price per Unit: ${Px(x_intercepts[0]):.2f}
   
    Maximum Total Revenue: ${Rx(x_intercepts[0]):,.2f}
""")
print("="*70)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def Px(x):
    """Per unit price function
    Input: x = total number of units
    Output: price of each unit ($)
    """
    return -(1/3)*x + 100

def Rx(x):
    """Total revenue function
    Input: x = total number of units
    Output: total revenue ($)
    """
    return -(1/3)*x**2 + 100*x

def dRx_dx(x):
    """Slope of tangent line to revenue function
    This is the derivative of Rx
    """
    return -(2/3)*x + 100

print("="*70)
print("MAXIMIZING REVENUE USING TANGENT SLOPE = 0 AND INEQUALITY CONCEPT")
print("="*70)

print("\n GIVEN FUNCTIONS:")
print("   Px(x) = -(1/3)x + 100     [Price per unit function]")
print("   Rx(x) = -(1/3)x + 100x   [Total revenue function]")
print("\n   Where x = total number of units")

# ============================================================================
# STEP 1: FIND VERTEX USING TANGENT SLOPE = 0
# ============================================================================
print("\n" + "="*70)
print("STEP 1: FINDING VERTEX WHERE TANGENT LINE SLOPE = 0")
print("="*70)

print("\n The revenue function Rx(x) is a DOWNWARD OPENING PARABOLA")
print("   At the vertex (maximum point), the tangent line is HORIZONTAL")
print("   This means: Slope of tangent line = 0")

print("\n Taking derivative of Rx(x):")
print("   Rx(x) = -(1/3)x + 100x")
print("   dRx/dx = -(2/3)x + 100")
print("\n   The derivative dRx/dx represents the SLOPE of the tangent line")

print("\n  Setting tangent slope = 0 to find vertex:")
print("   -(2/3)x + 100 = 0")
print("   -(2/3)x = -100")
print("   x = -100 / -(2/3)")
print("   x = 100  (3/2)")

# Calculate vertex by solving: -(2/3)x + 100 = 0
# Rearranging: x = 100 / (2/3)
x_vertex = 100 / (2/3)
print(f"   x = {x_vertex}")
price_vertex = Px(x_vertex)
revenue_vertex = Rx(x_vertex)

print(f"\n VERTEX FOUND AT x = {x_vertex} units")
print(f"   Price per unit at vertex: Px({x_vertex}) = ${price_vertex:.2f}")
print(f"   Total revenue at vertex: Rx({x_vertex}) = ${revenue_vertex:.2f}")

# Verify slope is actually 0
slope_at_vertex = dRx_dx(x_vertex)
print(f"\n Verification: Slope at x = {x_vertex} is {slope_at_vertex} (HORIZONTAL TANGENT)")

# ============================================================================
# STEP 2: VERIFY MAXIMUM USING INEQUALITY CONCEPT
# ============================================================================
print("\n" + "="*70)
print("STEP 2: VERIFY MAXIMUM USING INEQUALITY CONCEPT")
print("="*70)

print("\n Testing revenue at different points using INEQUALITY:")
print("   We'll check if Rx(vertex) > Rx(other points)")
print("   If TRUE for all nearby points, then vertex is the MAXIMUM")

# Test multiple points around the vertex
test_values = [100, 120, 140, 150, 160, 180, 200]
print(f"\n{'x (units)':<12} {'Revenue Rx(x)':<18} {'Comparison to Vertex':<30}")
print("-" * 70)

for x_test in test_values:
    rev = Rx(x_test)
    if x_test == x_vertex:
        comparison = " VERTEX (MAXIMUM) "
        print(f"{x_test:<12} ${rev:<17,.2f} {comparison}")
    else:
        diff = revenue_vertex - rev
        comparison = f"Rx({x_vertex}) - Rx({x_test}) = ${diff:,.2f} > 0"
        print(f"{x_test:<12} ${rev:<17,.2f} {comparison}")

print("\n INEQUALITY VERIFICATION:")
print(f"   For ALL tested values of x  {x_vertex}:")
print(f"   Rx({x_vertex}) > Rx(x)")
print(f"   ${revenue_vertex:,.2f} > Rx(x)")
print(f"\n   This PROVES that x = {x_vertex} gives the MAXIMUM revenue!")

# ============================================================================
# STEP 3: ANALYZE TANGENT SLOPE BEHAVIOR
# ============================================================================
print("\n" + "="*70)
print("STEP 3: ANALYZE TANGENT SLOPE BEHAVIOR AROUND VERTEX")
print("="*70)

test_slopes = [
    (100, "Before vertex"),
    (150, "At vertex"),
    (200, "After vertex")
]

print(f"\n{'x (units)':<12} {'Slope dRx/dx':<15} {'Interpretation':<30}")
print("-" * 70)

for x_test, position in test_slopes:
    slope = dRx_dx(x_test)
    if slope > 0:
        interpretation = "Positive  Revenue INCREASING"
    elif slope < 0:
        interpretation = "Negative  Revenue DECREASING"
    else:
        interpretation = "Zero  MAXIMUM (horizontal tangent)"
    
    print(f"{x_test:<12} {slope:<15.2f} {interpretation:<30} ({position})")

print("\n SLOPE PATTERN:")
print("    Before vertex (x < 150): Slope > 0 (function going UP)")
print("    At vertex (x = 150): Slope = 0 (HORIZONTAL tangent)")
print("    After vertex (x > 150): Slope < 0 (function going DOWN)")
print("\n   The slope changes from POSITIVE  ZERO  NEGATIVE")
print("   This is the characteristic of a MAXIMUM point!")

# ============================================================================
# VISUALIZATION
# ============================================================================
print("\n" + "="*70)
print("GENERATING VISUALIZATION...")
print("="*70)

x_vals = np.linspace(0, 300, 1000)
Rx_vals = Rx(x_vals)
dRx_vals = dRx_dx(x_vals)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 11))

# Plot 1: Revenue Function with Tangent Line at Vertex
ax1.plot(x_vals, Rx_vals, 'b-', linewidth=2.5, label='Revenue Rx(x)')
ax1.plot(x_vertex, revenue_vertex, 'ro', markersize=15, 
         label=f'Vertex: x={x_vertex}, R=${revenue_vertex:,.0f}', zorder=5)

# Draw horizontal tangent line at vertex
tangent_x = np.linspace(x_vertex - 50, x_vertex + 50, 100)
tangent_y = np.full_like(tangent_x, revenue_vertex)
ax1.plot(tangent_x, tangent_y, 'r--', linewidth=2, 
         label='Tangent line (Slope = 0)', alpha=0.7)

# Mark test points
for x_test in [100, 140, 160, 200]:
    ax1.plot(x_test, Rx(x_test), 'go', markersize=8, alpha=0.6)

ax1.axvline(x_vertex, color='r', linestyle=':', alpha=0.4)
ax1.axhline(revenue_vertex, color='r', linestyle=':', alpha=0.4)
ax1.grid(True, alpha=0.3)
ax1.set_xlabel('Total Number of Units (x)', fontsize=13, fontweight='bold')
ax1.set_ylabel('Total Revenue ($)', fontsize=13, fontweight='bold')
ax1.set_title('Revenue Function: Rx(x) = -(1/3)x + 100x (Downward Parabola)', 
              fontsize=15, fontweight='bold')
ax1.legend(fontsize=11, loc='upper right')
ax1.text(x_vertex + 10, revenue_vertex - 400, 
         f'MAXIMUM REVENUE\n\n' +
         f'Units: {x_vertex}\n' +
         f'Price: ${price_vertex:.2f}/unit\n' +
         f'Revenue: ${revenue_vertex:,.2f}',
         fontsize=11, bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8),
         verticalalignment='top')

# Plot 2: Tangent Slope Function
ax2.plot(x_vals, dRx_vals, 'g-', linewidth=2.5, label="Tangent Slope: dRx/dx = -(2/3)x + 100")
ax2.axhline(0, color='k', linestyle='-', linewidth=1)
ax2.plot(x_vertex, 0, 'ro', markersize=15, 
         label=f'Slope = 0 at x = {x_vertex} (Vertex)', zorder=5)
ax2.axvline(x_vertex, color='r', linestyle=':', alpha=0.4)

# Shade regions
ax2.fill_between(x_vals, 0, dRx_vals, where=(dRx_vals > 0), 
                 alpha=0.3, color='green', label='Slope > 0 (Revenue Increasing)')
ax2.fill_between(x_vals, 0, dRx_vals, where=(dRx_vals < 0), 
                 alpha=0.3, color='red', label='Slope < 0 (Revenue Decreasing)')

ax2.grid(True, alpha=0.3)
ax2.set_xlabel('Total Number of Units (x)', fontsize=13, fontweight='bold')
ax2.set_ylabel('Slope of Tangent Line', fontsize=13, fontweight='bold')
ax2.set_title('Tangent Slope Analysis: Slope = 0 at Maximum', 
              fontsize=15, fontweight='bold')
ax2.legend(fontsize=10, loc='upper right')
ax2.text(x_vertex, -8, f'Horizontal Tangent\n(Slope = 0)', 
         ha='center', fontsize=11, 
         bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8))

plt.tight_layout()
plt.savefig('revenue_maximization.png', dpi=300, bbox_inches='tight')
plt.show()

# ============================================================================
# FINAL ANSWER
# ============================================================================
print("\n" + "="*70)
print(" FINAL ANSWER: MAXIMUM REVENUE CONFIGURATION")
print("="*70)
print(f"""
To MAXIMIZE the total revenue:

    Total Number of Units to Produce/Sell: {x_vertex} units
   
    Price per Unit: ${price_vertex:.2f}
   
    Maximum Total Revenue: ${revenue_vertex:,.2f}

PROOF:
    Tangent slope = 0 at x = {x_vertex} (vertex of parabola)
    Inequality: Rx({x_vertex}) > Rx(x) for all x  {x_vertex}
    Slope changes: Positive  Zero  Negative (maximum pattern)
""")
print("="*70)
```

### Area Maximization with Parameter Constraint (Constant)
* ***3000 Meters of Fence you have (Constant)***
* ***By Fully using that Fence, You need to protect Maximum Land Area***
* ***Height = X, Length = Y***
* ***Constraint Equation --> 3000 = 2X + 2Y***
* ***Y Constraint --> Y = 1500 - X***
* ***Area = X * Y***
* ***Area Function Equation --> $A(x) = -x^2 + 1500x$***

```python
from sympy import symbols, solve
import sympy as smp

# Define the variable
x = symbols('x')

# Area Function
# Input is Height, Output is Area
# Downward Opening Parabola
Ax = -x**2 + 1500*x


# Tangent Slope Function (Derivative) of the Ax Function
dAx = smp.diff(Ax, x)


# Printing the Functions
print(f"Perimeter Constrant Equation: 3000 = 2X + 2Y")
print("Y Constrant Equation: Y = 1500 - X")
print(f"Total Area Function: A(x) = {Ax}")
print(f"Tangent Slope Function (Derivative) of A(x): dA(x) = {dAx}")
print()
print("Where x = Height as Input for A(x)")
print("Where x = Specific point x-coordinate, Height as Input for Tangent Slope Function dA(x)")


print("=" * 70)
print("SOLVING QUADRATIC INEQUALITIES")
print("=" * 70)


# 2. Solve A(x) >= 0
print("\n1. Inequalities Condition A(x) >= 0:")
solution1 = solve(Ax >= 0, x)
print(f"   Condition Solution: {solution1}")

print("\n" + "="*70)
print("STEP 2: FINDING VERTEX WHERE TANGENT LINE SLOPE = 0")
print("="*70)
print(f"Tangent Slope Function (Derivative) of A(x): dA(x) = {dAx}")


# 1. Find X-AXIS INTERCEPTS (where y = 0, slope m =0, i.e., dA(x) = 0)
x_intercepts = solve(dAx, x)
print(f"\nX-axis intercepts (roots): {x_intercepts}")
print(f"X-axis intercept point Coordinates are: ", end="")
for x_val in x_intercepts:
    print(f"({x_val}, 0)", end=" ")
print()


# Defining the functions seperately
def Ax(x):
    return -x**2 + 1500*x


# ============================================================================
# FINAL ANSWER
# ============================================================================
print("\n" + "="*70)
print(" FINAL ANSWER: MAXIMUM REVENUE CONFIGURATION")
print("="*70)
print(f"""
To MAXIMIZE the total Area:

   Optimum Height: {x_intercepts[0]} Feet
   
   Optimum Length: {1500 - x_intercepts[0]} Feet
   
   Maximum Total Area: {Ax(x_intercepts[0])} Feet**2
""")
print("="*70)
```

# Find out Tangent Slope Function (Slope Derivative) & Tangent Line Function at any point $(x_1, y_1)$ of the Original Function

**Tangent Slope Function** at any Point is also called **Slope Derivative Function**

**The Tangent Slope** Function is the derivative, $f'(x)$, which gives the steepness (slope) of the original function $f(x)$ at any point, representing its instantaneous rate of change.

**The Tangent Line** Function (or equation) is found using the point-slope formula $y - y_1 = m(x - x_1)$, where $m$ is the tangent slope $f'(x_1)$ and $(x_1, y_1)$ is the specific point on the curve, giving the equation for the straight line that just touches the curve at that point.

* **Tangent Slope Function** = **derivative** of the **original Function**

* By plugging in $x_1$ - coordinate of any point into Tangent Slope Function we will find out The Tangent Slope **(m)** of the **original function** at that specific point.

* For finding Tangent Line function of the original function at that specific point, we will use point-slope formula $y - y_1 = m(x - x_1)$, where $m$ is the tangent slope $f'(x_1)$ and $(x_1, y_1)$ is the specific point on the curve, giving the equation for the straight line that just touches the curve at that point.

```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt

# Define the symbol
x = smp.Symbol('x')

# Define the original function
# Example: f(x) = x^3 - 3x^2 + 2x + 1
f = x**3 - 3*x**2 + 2*x + 1

print("=" * 60)
print("TANGENT SLOPE AND LINE CALCULATOR")
print("=" * 60)
print(f"\nOriginal Function: f(x) = {f}")

# Step 1: Find the Tangent Slope Function (derivative) of the Original Function
f_prime = smp.diff(f, x)
print(f"\nTangent Slope Function: f'(x) = {f_prime}")

# Step 2: Choose a specific point x1
x1 = 3  # You can change this value

# Calculate y1 = f(x1)
y1 = f.subs(x, x1)
print(f"\nPoint on curve: ({x1}, {y1})")

# Step 3: Find the tangent slope at x1
m = f_prime.subs(x, x1)
print(f"\nTangent Slope at x = {x1}: m = f'({x1}) = {m}")

# Step 4: Find the Tangent Line Function using point-slope formula
# y - y1 = m(x - x1)
# y = m(x - x1) + y1
tangent_line = m * (x - x1) + y1
tangent_line_simplified = smp.expand(tangent_line)

print(f"\nTangent Line Function:")
print(f"  Point-slope form: y - {y1} = {m}(x - {x1})")
print(f"  Slope-intercept form: y = {tangent_line_simplified}")

# Visualization
print("\n" + "=" * 60)
print("VISUALIZATION")
print("=" * 60)

# Create lambdified functions for plotting
f_numeric = smp.lambdify(x, f, 'numpy')
tangent_numeric = smp.lambdify(x, tangent_line, 'numpy')

# Calculate y-intercept of tangent line (when x = 0)
y_intercept = tangent_line.subs(x, 0)
print(f"\nY-intercept of tangent line: (0, {y_intercept})")

# Generate x values for plotting
x_vals = np.linspace(x1 - 3, x1 + 3, 300)
y_vals = f_numeric(x_vals)
tangent_vals = tangent_numeric(x_vals)

# Create the plot
plt.figure(figsize=(12, 7))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label=f'f(x) = {f}')
plt.plot(x_vals, tangent_vals, 'r--', linewidth=2, 
         label=f"Tangent line at x={x1}: Slope(m)=f'({x1})={m}, Tangent Line eqn: y={tangent_line_simplified}")
plt.plot(x1, y1, 'go', markersize=10, label=f'Point of tangency: ({x1}, {y1})')
plt.plot(0, y_intercept, 'mo', markersize=10, label=f'Y-intercept: (0, {y_intercept})')

plt.axhline(y=0, color='k', linestyle='-', alpha=0.3, linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3, linewidth=0.5)
plt.grid(True, alpha=0.3)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title(f'Original Function and Tangent Line at ({x1}, {y1})', fontsize=14)
plt.legend(fontsize=9, loc='best')
plt.tight_layout()
plt.show()

print("\nPlot generated successfully!")

# Interactive function to calculate tangent at any point
def find_tangent_at_point(func, x_point):
    """
    Find tangent slope and tangent line at a specific point
    
    Parameters:
    func: sympy expression of the function
    x_point: x-coordinate of the point
    
    Returns:
    Dictionary with tangent slope and tangent line
    """
    x = smp.Symbol('x')
    
    # Derivative
    f_prime = smp.diff(func, x)
    
    # y-coordinate
    y_point = func.subs(x, x_point)
    
    # Tangent slope
    slope = f_prime.subs(x, x_point)
    
    # Tangent line
    tangent = slope * (x - x_point) + y_point
    tangent_simplified = smp.expand(tangent)
    
    return {
        'point': (x_point, y_point),
        'slope': slope,
        'tangent_line': tangent_simplified,
        'derivative': f_prime
    }

# Example usage of the function
print("\n" + "=" * 60)
print("TESTING WITH DIFFERENT POINT")
print("=" * 60)

result = find_tangent_at_point(f, 1)
print(f"\nAt x = 1:")
print(f"  Point: {result['point']}")
print(f"  Tangent Slope: {result['slope']}")
print(f"  Tangent Line: y = {result['tangent_line']}")
```

```python
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt

print("=" * 80)
print("GENERAL TANGENT LINE EQUATION FORMULA")
print("=" * 80)

# Define symbols
x = smp.Symbol('x')
x1 = smp.Symbol('x1')

# Define the original function
# Example: f(x) = x^3 - 3x^2 + 2x + 1
f = x**3 - 3*x**2 + 2*x + 1

print(f"\nOriginal Function:")
print(f"  f(x) = {f}")

# Step 1: Find the derivative f'(x)
f_prime = smp.diff(f, x)
print(f"\nStep 1: Find the derivative (Tangent Slope Function)")
print(f"  f'(x) = {f_prime}")

# Step 2: Substitute x1 into f(x) to get f(x1)
f_x1 = f.subs(x, x1)
print(f"\nStep 2: Evaluate f(x1)")
print(f"  f(x1) = {f_x1}")


# Step 3: Substitute x1 into f'(x) to get f'(x1)
# m = f'(x1)
# m = f_prime.subs(x, x1)
f_prime_x1 = f_prime.subs(x, x1)
print(f"\nStep 3: Evaluate f'(x1)")
print(f"  f'(x1) = {f_prime_x1}")


# Step 4: Apply the General Tangent Line Formula
print("\n" + "=" * 80)
print("GENERAL TANGENT LINE EQUATION FORMULA")
print("=" * 80)
print("\nFormula: y = f'(x1)  (x - x1) + f(x1)")
print("\nWhere:")
print("   f'(x1) = slope of tangent at point x1")
print("   (x1, f(x1)) = point of tangency")
print("   x = variable")


# Build the general tangent line equation symbolically
# General Tangent Line Equation: y = f'(x1)(x - x1) + f(x1)
# tangent_line_general = m * (x - x1) + y1
tangent_line_general = f_prime_x1 * (x - x1) + f_x1
print(f"\nSubstituting into the formula:")
print(f"  y = ({f_prime_x1})  (x - {x1}) + ({f_x1})")


# Expand the general equation
tangent_line_expanded = smp.expand(tangent_line_general)
print(f"\nExpanded General Tangent Line Equation:")
print(f"  y = {tangent_line_expanded}")


# Simplify and collect terms
tangent_line_simplified = smp.collect(tangent_line_expanded, x)
print(f"\nSimplified:")
print(f"  y = {tangent_line_simplified}")


print("\n" + "=" * 80)
print("APPLYING THE FORMULA TO SPECIFIC POINTS")
print("=" * 80)


# Below Portion is seperate part for calculations----------------------------------------------------------------------------------------


# Function to apply the general formula at specific points
def apply_tangent_formula(x_value):
    """
    Apply the general tangent line formula at a specific x value
    Using: y = f'(x1)  (x - x1) + f(x1)
    """
    # Substitute x1 = x_value into the general formula
    tangent_at_point = tangent_line_general.subs(x1, x_value)
    tangent_simplified = smp.expand(tangent_at_point)
    
    # Calculate specific values
    y_value = f.subs(x, x_value)
    slope_value = f_prime.subs(x, x_value)
    y_intercept = tangent_simplified.subs(x, 0)
    
    return {
        'x1': x_value,
        'y1': y_value,
        'slope': slope_value,
        'tangent_equation': tangent_simplified,
        'y_intercept': y_intercept
    }

# Test at multiple points
test_points = [0, 1, 2, 3]
results = []

for pt in test_points:
    result = apply_tangent_formula(pt)
    results.append(result)
    
    print(f"\nAt x1 = {pt}:")
    print(f"  Step 1: f({pt}) = {result['y1']}")
    print(f"  Step 2: f'({pt}) = {result['slope']}")
    print(f"  Step 3: Apply formula y = f'(x1)  (x - x1) + f(x1)")
    print(f"          y = {result['slope']}  (x - {pt}) + {result['y1']}")
    print(f"  Result: y = {result['tangent_equation']}")
    print(f"  Y-intercept: {result['y_intercept']}")

# Visualization
print("\n" + "=" * 80)
print("VISUALIZATION: GENERAL FORMULA APPLIED TO MULTIPLE POINTS")
print("=" * 80)

fig, ax = plt.subplots(figsize=(12, 8))

# Plot original function
x_range = np.linspace(-1, 4, 500)
f_numeric = smp.lambdify(x, f, 'numpy')
y_range = f_numeric(x_range)
ax.plot(x_range, y_range, 'b-', linewidth=3, label=f'f(x) = {f}', zorder=1)

# Colors for tangent lines
colors = ['red', 'green', 'orange', 'purple']

# Plot each tangent line
for i, result in enumerate(results):
    x_val = result['x1']
    y_val = result['y1']
    slope = result['slope']
    tangent_eq = result['tangent_equation']
    y_int = result['y_intercept']
    
    # Convert to numeric function
    tangent_numeric = smp.lambdify(x, tangent_eq, 'numpy')
    tangent_y = tangent_numeric(x_range)
    
    # Plot tangent line with formula in legend
    ## ax.plot(x_range, tangent_y, '--', color=colors[i], linewidth=2.5,
            ## label=f'x={x_val}: y = {slope}(x - {x_val}) + {y_val} = {tangent_eq}',
            ## alpha=0.8, zorder=2)

    # Plot tangent line with formula in legend (my code)
    ax.plot(x_range, tangent_y, '--', color=colors[i], linewidth=2,
            label=f"Tangent line at x={x_val}: Slope(m)=f'({x_val})={slope}, "
            f"Tangent Line eqn: y={tangent_eq}, Tangency Point = {x_val, y_val}, "
            f"Y-intercept = {0, y_int}", alpha = 0.8, zorder = 2)
    
    
    # Mark point of tangency
    ax.plot(x_val, y_val, 'o', color=colors[i], markersize=12,
            markeredgecolor='black', markeredgewidth=2, zorder=4)
    
    # Mark y-intercept
    ax.plot(0, y_int, 's', color=colors[i], markersize=10,
            markeredgecolor='black', markeredgewidth=1.5, alpha=0.7, zorder=3)
    
    # Add annotation showing the formula application
    offset = 0.3 * (i - 1.5)
    ax.annotate(f'({x_val}, {y_val})',
                xy=(x_val, y_val), xytext=(x_val + 0.5, y_val + offset),
                fontsize=9, color=colors[i], fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='white', 
                         edgecolor=colors[i], alpha=0.8),
                arrowprops=dict(arrowstyle='->', color=colors[i], lw=1.5))

# Add grid and axes
ax.axhline(y=0, color='k', linestyle='-', alpha=0.3, linewidth=1)
ax.axvline(x=0, color='k', linestyle='-', alpha=0.3, linewidth=1)
ax.grid(True, alpha=0.3, linestyle='--')

# Labels and title
ax.set_xlabel('x', fontsize=14, fontweight='bold')
ax.set_ylabel('y', fontsize=14, fontweight='bold')
ax.set_title('General Tangent Line Formula: y = f\'(x)(x - x) + f(x)\nApplied at Multiple Points',
             fontsize=16, fontweight='bold', pad=20)

# Legend
ax.legend(fontsize=9, loc='upper left', framealpha=0.95, 
          title='Tangent Lines ( = tangency point,  = y-intercept)',
          title_fontsize=10)

ax.set_xlim(-1, 4)
ax.set_ylim(-8, 10)

plt.tight_layout()
plt.show()

# Summary box
print("\n" + "=" * 80)
print("FORMULA SUMMARY")
print("=" * 80)
print("\nGeneral Tangent Line Equation Formula:")
print("" + "" * 78 + "")
print("  y = f'(x)  (x - x) + f(x)                                             ")
print("" + "" * 78 + "")
print("\nFor our function f(x) = " + str(f) + ":")
print(f"   Derivative: f'(x) = {f_prime}")
print(f"   General Formula: y = {tangent_line_simplified}")
print("\nTo find tangent at any specific point x:")
print("  1. Calculate f(x) - the y-coordinate")
print("  2. Calculate f'(x) - the slope")
print("  3. Substitute into: y = f'(x)  (x - x) + f(x)")
print("  4. Simplify to get the tangent line equation")

print("\n" + "=" * 80)
```
